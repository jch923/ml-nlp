### 데이터 분석 시 시각화를 위한 라이블러리

### matplotlib.pyplot
#### 파이플롯은 맷플롯립안에 포함된 서브 모듈이다.
#### 간단한 시각화의 경우 파이플롯만 사용하더라도 충분하다.

```python
# 파이플롯 사용을 위해 불러온다.
import matplotlib.pyplot as plt
import numpy as np
```

```python
# 주피터 노트북 내부에서 그림 표시를 위해 선언해준다.
%matplotlib inline  
```

### 기본적인 그래프 그리기

```python
# 임의의 그래프 생성
x = [1,3,5,7,9]
y = [100,200,300,400,500]
plt.plot(x, y)
```

### 코사인 함수 그리기

```python
# 연속적인 값을 갖는 배열
x = np.linspace(-np.pi, np.pi, 128) 
# x 리스트에 대한 cos값 계산
y = np.cos(x) 
plt.plot(y)
```

### 판다스 데이터 시각화
#### 데이터는 보통 판다스를 사용해서 다루기 때문에 판다스 데이터를 시각화 하는 것이 중요하다.

```python
import pandas as pd
# 데이터를 읽어온다.
data_frame = pd.read_csv('./data_in/datafile.csv') 
# plot을 이용하여 데이터 시각화
data_frame.plot()

# C값만 보이는 이유는 데이터 값이 상대적으로 다른값에 비해 크기 떄문이다.
```

```python
# 누적된 값을 가지고 데이터의 증감에 대해 시각화 하고 싶다면 Pandas.cumsum()함수를 이용한 후 시각화 하면 된다.
#데이터를 누적값으로 바꿔준다.
data_sum = data_frame.cumsum() 
data_sum.plot()
```

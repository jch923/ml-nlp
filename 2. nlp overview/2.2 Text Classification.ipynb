# 텍스트 분류

텍스트 분류는 자연어 처리 기술을 활용해 특정 텍스트를 사람들이 정한 몇가지 범주(Class)중 어느 범주에 속하는지 분류하는 문제다.
분류해야 할 범주의 수에 따라 문제를 구분하기도 하는데 보통 2가지 범주에 대해 구분하는 문제를 이진 분류 문제라 한다.
그리고 3개 이상의 범주에 대해 분류하는 문제를 통틀어 다중 범주 분류 문제라 한다.

## 텍스트 분류의 예시

### 스팸 분류

일반 메일과 스팸 메일을 분류하는 문제가 스팸 분류 문제다.
여기서 분류해야 할 범주(class)는 스팸 메이로가 일반 메일 2가지이다.

### 감정 분류

감정 분류란 주어진 글에 대해 이 글이 긍정적인지 부정적인지 판단하는 문제다.
이 경우 범주는 긍정 혹은 부정이 된다.
경우에 따라서 중립의 범주가 추가될 수도 있고 범주를 세분화 할 수도 있다.
문제에서 분류하려는 의도에 따라 범주가 정해지는 것이다.
가장 대표적인 감정 분류 문제는 영화 리뷰에 대해 각 리뷰가 긍정적인지 부정적인지 판단하는 문제인데, 다음 장에서 직접 해볼 예정이다.

### 뉴스 기사 분류

뉴스 기사는 사용자가 원하는 뉴스를 선택해서 볼 수 있게 범주를 잘 구분지어 분류되어 있어야 한다.
자연어 처리 기술을 이용하면 이를 더 효율적으로 분류할 수 있다.

위의 예시들 뿐만 아니라 텍스트 분류에는 다양한 문제가 있다.
분류하는 단위를 글 전체로 하지 않고 각 단어를 한단위로 분류하는 문제도 있다.
이러한 텍스트 분류 문제는 지도학습, 비지도학습을 통해 해결할 수 있다.

## 지도 학습을 통한 텍스트 분류

지도 학습은 글(데이터)에 대해 각각 속한 범주에 대한 값(라벨)이 이미 주어져 있다.
따라서 주어진 범주로 글들을 모두 학습한 후 학습한 결과를 이용해 새로운 글의 범주를 예측하는 방법이다.

예를들어 스팸 분류 문제를 지도학습으로 해결한다고 하자면,
10000개의 메일 데이터가 있고, 이중 5000개는 스팸메일, 5000개는 일반 메일이라고 가정한다.
각 메일에는 해당 메일이 어떤 메일인지 라벨링 되어있다.
이제 각 메일에서 특징을 뽑아내서 예측한 뒤 라벨과 맞는 지 확인하면서 학습한다.
10000개의 메일을 모두 사용해서 학습한 뒤 이제 새로운 메일에 대해 삭습한 모델을 통해 스팸 메일인지 아닌지 예측한다.

## 비지도학습을 통한 텍스트 분류

비지도 학습에서는 데이터만 존재하고, 각 데이터는 범주로 미리 나눠져 있지 않다.
따라서 특성을 찾아내서 적당한 범주를 만들어 각 데이터를 나누면 된다.
비지도 학습을 통한 분류는 어떤 특정한 분류가 있는 것이 아니라 데이터의 특성에 따라 비슷한 데이터끼리 묶어주는 개념이다.
이렇게 묶어준 각 묶음이 하나의 범주가 되는 것이다.

지도학습과 비지도학습 중 어떤 방법을 사용할지 결정하는 데 기여하는 가장 큰 기준은 데이터에 정답 라벨이 있느냐 없느냐다.
정답 라벨이 있는 경우 지도학습 방법, 없다면 비지도 학습을 사용하면 된다.
